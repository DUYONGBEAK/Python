```python
import requests
# xml 파일을 읽어오기 위해 필요함
from lxml import html
# 태그형태의 데이터를 추출하기 위해 필요함
from bs4 import BeautifulSoup
import pandas as pd
# 아래 두줄은 url인코더를 하기 위해 필요함
from urllib.request import Request, urlopen
from urllib.parse import urlencode, quote_plus, unquote

# API불러오기 (두가지 방법)

# API_Key = unquote('key')
# url = ""
# queryParams = '?' + urlencode(
# {
# quote_plus('ServiceKey') : API_Key,
# quote_plus('LAWD_CD'): '11110',
# quote_plus('DEAL_YMD'): '201512'
# }
# )
# response = requests.get(url+queryParams)
response = requests.get("http://openapi.molit.go.kr/OpenAPI_ToolInstallPackage/service/rest/RTMSOBJSvc/getRTMSDataSvcAptTradeDev?serviceKey=AT33s775KYpJOkUBJu0dxkJuUeIfDIOJRzAH084EQS3JN%2BzFjErLHuk%2FGZa9L4gBTSGCzeA69tI9PwLp7B37IQ%3D%3D&pageNo=1&numOfRows=10&LAWD_CD=11110&DEAL_YMD=201512")

# 인코딩하기
response.encoding = 'utf-8'

# parser
xmlobj = BeautifulSoup(response.text,"lxml-xml")

# item태그안에 있는 내용 찾기
rows = xmlobj.find_all("item")

# 사용할 배열 초기화
rowList = []
columnList = []
columnsNameList = []

# 반복문을 통해 데이터를 배열에 입력
for i in range(0, len(rows)):
    
    eachRow = rows[i]
    columns = eachRow.find_all()

    for j in range(0, len(columns)):
        eachColumns = columns[j]
        if i == 0:
            # name은 이름이 나오는것
            columnsNameList.append(eachColumns.name)
        # text내용이나옴
        eachColumnsValue = eachColumns.text
        
        columnList.append(eachColumnsValue.replace(" ",""))
        
    rowList.append(columnList)
    # 배열 초기화해서 깨끗하게 만들기
    columnList = []

indata = pd.DataFrame(rowList, columns=columnsNameList)

indata.head()
```
