```python
# 상대적 경로 지정
targetData = ("../../stddev/dataset/kopo_product_volume.csv")

#데이터 인코딩 후 읽어오기
data = pd.read_csv(targetData, encoding="ms949")

#컬럼명 변경
data.columns = ["regiond","productgroup","yearweek","volume"]

# DB에 접속하기 위한 라이브러리
import psycopg2
from sqlalchemy import create_engine

# postgres접속 (DB이름://ID:PW@IP주소/DB이름)
mylocalpgUrl = "postgresql://postgres:1234@127.0.0.1/postgres"

# 주소로 접속하기 위한 엔진처리
engine_pg = create_engine(mylocalpgUrl)

# 테이블 명을 변수로 지정
resultTableName = "pg_out"

# PC에 저장되어 있는 파일을 DB에 보내는 과정
data.to_sql(name =resultTableName,
           con = mylocalpgUrl, if_exists="replace", index = False)

# DB에 저장되어 있는 sql을 읽어오는 방법1
pd.read_sql(sql="select * from pg_out",con = engine_pg)

# DB에 저장되어 있는 sql을 읽어오는 방법2
pd.read_sql_table(table_name="pg_out", con = engine_pg)
```

```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine

# DB 경로 지정
otherpgUrl = "postgresql://postgres:postgres@192.168.110.22/postgres"

# 주소로 접속하기 위한 엔진처리
practice = create_engine(otherpgUrl)

# 타인 IP주소에 들어있는 DB에 접속해서 sql을 읽어옴
dataData = pd.read_sql(sql = "select * from pg_out", con = practice)

# 컬럼명을 변경
dataData.columns = ["salesid","productgroup","yearweek","volume"]

# 변경된 컬럼명을 상대방 DB에 전송
dataData.to_sql(name = "pg_out", con = practice, 
                   if_exists="replace", index = False)
```
