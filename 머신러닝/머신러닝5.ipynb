```python
# 머신러닝, 타임시리즈 모델, 딥러닝
```

```python
import statsmodels.formula.api as smf
```

```python
ml = smf.logit(formula="diabetes ~ glucose+bmi+age+insulin",data=indianData).fit()
```

```python
ml.summary()
```

```python
features = ["glucose","bmi","age","insulin"]
```

```python
label = ["diabetes"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainingDataFeatures,\
testDataFeatures,\
trainingDatalabel,\
testDatalabel=\
train_test_split(indianData.loc[:,features],\
                indianData.loc[:,label], random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_method = LogisticRegression()
```

```python
modelFit = model_method.fit(trainingDataFeatures,trainingDatalabel)
```

```python
predictValue = modelFit.predict(testDataFeatures)
```

```python
testDataIndex = testDataFeatures.index
```

```python
testDataAll = indianData.loc[testDataIndex]
```

```python
testDataAll["predict"] = predictValue
```
