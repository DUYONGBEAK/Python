```python
import pandas as pd
import numpy as np

# 데이터 불러오기
selloutData = \
pd.read_csv("../../stddev/dataset/kopo_channel_seasonality_new.csv", encoding = "ms949")

# QTY의 음수를 제거하는 함수
def realQTY(inValue):
    if float(inValue) < 0:
        outValue = 0
    elif float(inValue) >= 0:
        outValue = inValue
    return outValue    

# 음수QTY를 새로운 컬럼에 맵핑함
selloutData["QTY_NEW"] = selloutData.QTY.apply(realQTY)

# 음수QTY가 제거 됐는지 확인하는 로직검증
selloutData.loc[ (selloutData.QTY < 0) &
                 (selloutData.QTY_NEW != 0)]

# 모든 컬럼을 str로 변환함
selloutData = selloutData.astype( { "REGIONID":str,
                    "PRODUCT":str,
                    "YEARWEEK":str,
                    "QTY":str,
                    "QTY_NEW":str               
                    } )

# YEARWEEK를 YEAR와 WEEK로 분리함
selloutData["YEAR"] = selloutData.YEARWEEK.str[0:4]
selloutData["WEEK"] = selloutData.YEARWEEK.str[-2:]

# WEEK중에서 53을 제거함
selloutData.WEEK = np.where(selloutData.WEEK == "53", "00",
         np.where(selloutData.WEEK == "00", "00", selloutData.WEEK))

# 결과검증
selloutData.loc[selloutData.WEEK == "53"]

# 결과 출력
refinedSelloutData = selloutData.loc[selloutData.WEEK != "00"]
refinedSelloutData
```
